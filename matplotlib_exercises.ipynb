```python
import matplotlib.pyplot as plt
```

```python
x = range(-10, 11)
y = [ (n ** 2 - n + 2) for n in x]
```

```python
plt.plot(x, y)
plt.text(0, 0, '(0, 0)', fontsize=10, color='blue', ha='center')
plt.show()
```

```python
import math
import numpy

x = range(1, 100)
y = [math.sqrt(n) for n in x]

plt.text(40, 8, '$y= sqrt(x)$', fontsize=10, color='black')
plt.plot(x, y)
plt.show()
```

```python
x = range(-10, 10)
y = [n**3 for n in x]

plt.text(.25, 50, '$y= x^3$', fontsize=10, color='black')
plt.plot(x, y)
plt.show()
```

```python
x = range(0, 10)
y = [2**n for n in x]

plt.text(.25, 50, '$y= 2^x', fontsize=10, color='black')
plt.plot(x, y)
plt.show()
```

```python
x = range(0, 100)
y = [(1/(n+1)) for n in x]

x2 = range(-100,-1)
y2 = [(1/(n+1)) for n in x2]

plt.text(-6, 0, '$y = 1/(x+1)$', fontsize=10, color='black')
plt.plot(x, y)
plt.plot(x2, y2)
plt.show()
```

```python
plt.subplot(2,2,1)
x = range(1, 100)
y = [math.sqrt(n) for n in x]
plt.plot(x, y)

plt.subplot(2,2,2)
x = range(-10, 10)
y = [n**3 for n in x]

plt.subplot(2,2,3)
x = range(0, 10)
y = [2**n for n in x]
plt.plot(x, y)

plt.subplot(2,2,4)
x = range(0, 100)
y = [(1/(n+1)) for n in x]
x2 = range(-100,-1)
y2 = [(1/(n+1)) for n in x2]
plt.plot(x, y)
plt.plot(x2, y2)

plt.show()
```
