```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
import matplotlib.image as mpimg

x = range(0, 25)
y = [0*n+1 for n in x]

plt.plot(x,y, label = '0(1)', c='darkviolet', linewidth='4')

x = range(1, 25)
y = [math.log(n) for n in x]
plt.plot(x,y, label = '0(log n)', c='darkturquoise', linewidth='4')

x = range(0, 25)
y = [n for n in x]
plt.plot(x,y, label = '0(n)', c='orangered', linewidth='4')

x = range(1, 25)
y = [n*math.log(n) for n in x]
plt.plot(x,y, label = '0(n log n)', c='yellow', linewidth='4')

x = range(0, 25)
y = [n**2 for n in x]
plt.plot(x,y, label = '0$n^2$', c='cyan', linewidth='4')

x = range(0, 25)
y = [2**n for n in x]
plt.plot(x,y, label = '0$2^n$', c='aquamarine', linewidth='4')

x = range(0, 25)
y = [math.perm(n) for n in x]
plt.plot(x,y, label = '0(n!)', c='limegreen', linewidth='4')

x = range(0, 25)
y = [n**n for n in x]
plt.plot(x,y, label = '0$n^n$', c='hotpink', linewidth='4')

plt.style.use('dark_background')
plt.ylim(0, 100)
plt.xticks(fontsize='12', weight='bold')
plt.yticks(fontsize='12', weight='bold')
plt.legend(loc='upper right', ncol=2, fontsize='small', shadow='true')
plt.xlabel('Elements', fontsize='12', weight='bold')
plt.ylabel('Operations', fontsize='12', weight='bold', rotation=90)
plt.title('Big O Notation', fontsize='12', weight ='bold')


plt.show()
```
